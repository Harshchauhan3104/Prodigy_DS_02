```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
train_df = pd.read_csv('C:\\Users\\harsh\\OneDrive\\Desktop\\train.csv')
test_df = pd.read_csv('C:\\Users\\harsh\\OneDrive\\Desktop\\test.csv')
```

```python
# Check for missing values
print(train_df.isnull().sum())

# Fill missing values in 'Age' with the median age
train_df['Age'].fillna(train_df['Age'].median(), inplace=True)

# Fill missing values in 'Embarked' with the mode (most frequent value)
train_df['Embarked'].fillna(train_df['Embarked'].mode()[0], inplace=True)

# Drop the 'Cabin' column due to too many missing values
train_df.drop(columns=['Cabin'], inplace=True)
```

```python
# Convert 'Sex' to numerical values: male=0, female=1
train_df['Sex'] = train_df['Sex'].map({'male': 0, 'female': 1})

# Convert 'Embarked' to numerical values: S=0, C=1, Q=2
train_df['Embarked'] = train_df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
# Basic statistics of the dataset
print(train_df.describe())
```

```python
# Distribution of Age
sns.histplot(train_df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
# Survival rate by gender
sns.barplot(x='Sex', y='Survived', data=train_df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Survival rate by passenger class
sns.barplot(x='Pclass', y='Survived', data=train_df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(train_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
